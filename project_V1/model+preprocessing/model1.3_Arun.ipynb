```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
#from sklearn.tree import DecisionTreeClassifier
#from sklearn.neighbors import KNeighborsClassifier
#from sklearn.svm import LinearSVC
#from sklearn.ensemble import RandomForestClassifier
#from sklearn.model_selection import cross_val_score
#from sklearn.metrics import confusion_matrix
from sklearn import metrics
import re
import pickle
```

```python
st=["it is a very sad day today", " today is a boring day ", "wake up with determination go tp bed with satisfaction","my friend has been diagnosed with depression"," what the fuck"]
```

```python
ok = pd.read_csv("emotions_preprocessed_data.csv")
```

```python
ok1=ok['content']
```

```python
def remove_non_ascii(df):
    #s2 = str(df) 
    s2 = df.str.replace("[^\\x00-\\x7F]", "abcde ")
    return s2

ok1=remove_non_ascii(ok1)
```

```python
def remove_newline(df):
     #s2 = str(df) 
     s2= df.str.replace("\r\n", " ")
     return s2
   
ok1=remove_newline(ok1)
```

```python
def remove_punctuation(df):
    #s2 = str(df) 
    s2=df.str.replace('[^\w\s]',' ')
    return s2
   
ok1=remove_punctuation(ok1) 
```

```python
ok['content']=ok1
```

```python
ok.reset_index(drop=True)
```

```python
t_ok=[]
l_ok=[]


def remove_emoticon_ok_replacement(s):
    for i in range(len(s['content'])):
        x = re.search("(abcde)",s.loc[i]['content'])
        if (x):
            continue
        else:
            t_ok.append(s.loc[i]['content'])
            l_ok.append(s.loc[i]['sentiment']) 
ok=remove_emoticon_ok_replacement(ok) 
```

```python
s_ok=pd.DataFrame(columns=['text','label'])
```

```python
s_ok['text']=t_ok
s_ok['label']=l_ok
```

```python
count_vector = CountVectorizer()# define CountVectorizer
dataframe_count_ok = count_vector.fit_transform(s_ok['text'])#fit the vector

vectorizer = TfidfVectorizer()
A = vectorizer.fit_transform(s_ok['text']).toarray()

#d1=count_vector.fit_transform(mylist)


tfidf_transform = TfidfTransformer()# define Tfidf Transformer
dataframe_count_tfidf_ok = tfidf_transform.fit_transform(dataframe_count_ok)



with open("x_result.pkl", 'wb') as handle:pickle.dump(vectorizer, handle)
```

```python
vectorizer = pickle.load(open("x_result.pkl", "rb" ) )
B_ok=vectorizer.transform(st).toarray()
B_ok.shape
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(dataframe_count_tfidf_ok,s_ok['label'].astype(int),test_size=0.3 ,random_state = 23)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0, solver='lbfgs',
                 multi_class='multinomial').fit(X_train1, y_train1)
y_pred=clf.predict(X_test1)
accuracy_score(y_test1, y_pred)
```
